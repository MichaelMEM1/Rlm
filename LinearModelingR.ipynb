## How to read a CSV in R : read.csv() and place the csv file in quotes.

```r
df = read.csv("regrex1.csv")
```

## Reading the data

```r
head(df) 
```

## Ploting the data

```r
plot(df$x,df$y, xlab = 'X', ylab = 'Y') 
title('Plot of Data')
legend('topleft',                          # location of the legend
       legend=c('Points'),                 # labels for the lines
       col=c('black'),                 # colors matching the lines
         pch = 1,                        # Adding a point to the legend
       title='Legend',                        # legend title
       bty='n')                              # no box around the legend
```

 ## Applying the linear model to df 

```r
model = lm(y~x, data = df )
```

## Creating a variable for the statistics

```r
summary_stats = summary (model) 
```

## Printing the statistics

```r
summary_stats 
```

## Creating a variable for the r squared value and printing it 

```r
r_squared <- summary_stats$r.squared 

r_squared 
```

## Printing the x and y intercept

```r
coef(model)
```

## Selecting and printing the y intercept

```r
coef(model)[1] 
```

## selecting and printing the x intercept

```r
coef(model)[2] 
```

## Predicting the trend based on df dataframe 

```r
y_predict = predict(model, df) 
```

## Plotting the fit

```r
plot(df$x,y_predict, xlab = 'X', ylab = 'Y Prediction') 
title ('Predicted y values')
legend('topleft',                          # location of the legend
       legend=c('Points'),                 # labels for the lines
       col=c('black'),                 # colors matching the lines
         pch = 1,                        # Adding a point to the legend
       title='Legend',                        # legend title
       bty='n')                              # no box around the legend
```

## Plotting the data with the fit line 

```r
plot(df$x,df$y, xlab = 'X', ylab ='Y', col = 'black') 
lines(df$x,y_predict, col = 'red')
title('Linear regression model R')
legend('topleft',                          # location of the legend
       legend=c('Points', 'Fit'),                 # labels for the lines
       col=c('black', 'red'),                 # colors matching the lines
       lwd=c(2, 2),                          # line widths
       title='Legend',                        # legend title
       bty='n')                              # no box around the legend
text(12, 12, sprintf("R^2 = % f", r_squared), pos = 3, ) # adding the r square value
```
