```r
df = read.csv("regrex1.csv")# How to read a CSV in R : read.csv() and place the csv file in quotes.
```

```r
head(df) #reading the data
```

```r
plot(df$x,df$y, xlab = 'X', ylab = 'Y') # ploting the data
title('Plot of Data')
legend('topleft',                          # location of the legend
       legend=c('Points'),                 # labels for the lines
       col=c('black'),                 # colors matching the lines
         pch = 1,                        # Adding a point to the legend
       title='Legend',                        # legend title
       bty='n')                              # no box around the legend
```

```r
model = lm(y~x, data = df ) # applying the linear model to df 
```

```r
summary_stats = summary (model) # creating a variable for the stats
```

```r
summary_stats # printing the stats
```

```r
r_squared <- summary_stats$r.squared #creating a variable for the r squared value

r_squared # printing the r squared value
```

```r
coef(model) # printing the x and y intercept
```

```r
coef(model)[1] # selecting and printing the y intercept
```

```r
coef(model)[2] # selecting and printing the x intercept
```

```r
y_predict = predict(model, df) # predicting the trend based on df dataframe 
```

```r
plot(df$x,y_predict, xlab = 'X', ylab = 'Y Prediction') # plotting the fit
title ('Predicted y values')
legend('topleft',                          # location of the legend
       legend=c('Points'),                 # labels for the lines
       col=c('black'),                 # colors matching the lines
         pch = 1,                        # Adding a point to the legend
       title='Legend',                        # legend title
       bty='n')                              # no box around the legend
```

```r
plot(df$x,df$y, xlab = 'X', ylab ='Y', col = 'black') # plotting the data with the trend line
lines(df$x,y_predict, col = 'red')
title('Linear regression Model R')
legend('topleft',                          # location of the legend
       legend=c('Points', 'Fit'),                 # labels for the lines
       col=c('black', 'red'),                 # colors matching the lines
       lwd=c(2, 2),                          # line widths
       title='Legend',                        # legend title
       bty='n')                              # no box around the legend
text(12, 12, sprintf("R^2 = % f", r_squared), pos = 3, ) # adding the r square value
```
